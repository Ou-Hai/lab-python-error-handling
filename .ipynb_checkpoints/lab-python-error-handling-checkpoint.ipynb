def initialize_inventory(products):

   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity (integer).")
    return inventory


def calculate_total_price(orders):
    """
    orders 是一个字典：{product: quantity}
    在这里向用户询问每种产品的单价，并计算总价。
    """
    total_price = 0.0
    prices = {}

    for product, quantity in orders.items():
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price for each {product}: "))
                if price < 0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                valid_price = True
                prices[product] = price
            except ValueError as error:
                print(f"Error: {error}")

        total_price += price * quantity

    print("\nPrices used:", prices)
    return total_price


def get_customer_orders(inventory):

    orders = {}


    while True:
        try:
            num_orders = int(input("Enter the number of different products in the order: "))
            if num_orders < 0:
                print("Number of orders cannot be negative. Please enter a valid number.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter a valid integer for the number of orders.")

  
    for i in range(num_orders):
        print(f"\nOrder #{i + 1}")
        # 2.1 选择产品名称（带错误处理）
        while True:
            try:
                product = input("Enter the product name: ").strip().lower()
                if product not in inventory:
                    raise ValueError("Product not found in inventory.")
                if inventory[product] <= 0:
                    raise ValueError("This product is out of stock.")
                break
            except ValueError as error:
                print(f"Error: {error} Please enter a valid product name.")

        
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product}s to order: "))
                if quantity <= 0:
                    print("Quantity must be positive. Please enter a valid quantity.")
                elif quantity > inventory[product]:
                    print(f"Not enough stock. Available: {inventory[product]}. Please enter a smaller quantity.")
                else:
                    break
            except ValueError:
                print("Invalid input. Please enter a valid integer quantity.")

        orders[product] = orders.get(product, 0) + quantity
        inventory[product] -= quantity

    return orders



def main():
    products = ["apple", "banana", "orange"]

    print("=== Step 1: Initialize inventory ===")
    inventory = initialize_inventory(products)
    print("\nCurrent inventory:", inventory)

    print("\n=== Step 2 & 3: Get customer orders and calculate total price ===")
    orders = get_customer_orders(inventory)
    print("\nCustomer orders:", orders)

    total = calculate_total_price(orders)
    print(f"\nTotal price for this order: {total}")

    print("\nRemaining inventory:", inventory)


if __name__ == "__main__":
    main()
````
